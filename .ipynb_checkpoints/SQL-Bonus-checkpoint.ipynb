```python
import pandas as pd
import os
import logging
import pandas.io.sql as psql
import sqlalchemy
import matplotlib.pyplot as plt
from config import username, password
%matplotlib inline 
```

```python
from sqlalchemy import create_engine
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/animals.db')
connection = engine.connect()
```

```python
df = pd.read_sql("SELECT title, AVG(salary) AS Average_Salary FROM salary_emp as s NATURAL JOIN title_emp as t GROUP BY title;", connection)
```

```python
print(df)
```

```python
 ax = df.plot.bar(x='title', y='average_salary', title ='Average Salary of Employees', rot=20)
plt.savefig("Images/salary_barchart.png")
```

```python
s_df = pd.read_sql("SELECT salary, emp_no FROM salary_emp", connection)
```

```python
bx=s_df.plot.hist(bins=20, x= 'emp_no', title= 'Salary Histogram')
plt.savefig("Images/salary_hist.png")
```
